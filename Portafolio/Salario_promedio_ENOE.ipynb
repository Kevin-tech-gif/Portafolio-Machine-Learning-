Objetivo es explicar los cambios en el salario mensual a traves de variables sociodemograficas y laborales.

*Variables independientes: dummy de sexo, a√±os de escolaridad aprobada, dummy estado civil, horas laboradas al mes, experiencia laboral, dummy que indique el tama√±o de la localidad.

Creamos la dummmy de mujer


```python
dummy_sex = pd.get_dummies(subset["sex"], prefix="sex")
```

```python
dummy_sex
```

```python
dummy_sex = dummy_sex.rename(columns={'sex_1': 'hombre'})
dummy_sex = dummy_sex.rename(columns={'sex_2': 'mujer'})
dummy_sex
```

```python
data = pd.concat([subset, dummy_sex], axis = 1)
```

```python
data
```

generar una variable con estas caracteristicas: urbano=1 localidad cuyo tama√±a de poblacion son mayores a 2499 habitantes urbano=0 localidad cuyo tama√±o de poblacion es menor a 2500 habitantes (zona rural)

#Tipo de localidad
Comprende a cuatro grandes √°reas de poblaci√≥n: √Åreas m√°s urbanizadas: de 100 000 y m√°s habitantes. Urbano medio: de 15 000 a 99 999 habitantes. Urbano bajo: de 2 500 a 14 999 habitantes. Rural: de menos de 2 500 habitantes.

1	√Åreas m√°s urbanizadas

2	Urbano medio

3	Urbano bajo

4	Rural


```python
tabulacion = data["t_loc"].value_counts()
tabulacion
```

```python
import numpy as np
```

```python
data['urbano'] = np.where(data['t_loc'].isin([1, 2, 3]), 1, 0)
```

```python
data
```

```python
a√±os_ecolaridad_tab = data["anios_esc"].value_counts().sort_index()
a√±os_ecolaridad_tab
```

eliminamos el 99 en los a√±os de escolaridad que nos indica de las personas que no dieron respuesta a esa variable en particular


```python
data = data[data['anios_esc'] != 99]
```

```python
data
```

contrimos una dummy para el estado civil **Dummy_casado=1 si la persona es casado y 0 con personas con un estado civil distinto al casado y los que no sabe

```python
estado_conyugal_tab = data["e_con"].value_counts().sort_index()
estado_conyugal_tab
```

```python
data['casado'] = data['e_con'].apply(lambda x: 1 if x == "5" else 0)
```

```python
data
```

generamos las horas que trabaja una persona al mes ya que las horas de la variable hrsocup estan semanales y las ocupamos mensual


```python
horas_trabajadas = data["hrsocup"].value_counts().sort_index()
horas_trabajadas
```

```python
data["horas_mes"] = data["hrsocup"]*4
```

```python
data
```

esperiencia laboral

ecuacion de mincer


```python
data['eda'] = data['eda'].astype(int)
```

```python
data.info()
```

```python
data["exp_laboral"] = data["eda"] - data["anios_esc"] - 6
```

```python
data
```

#regresion del explicacion del salario real

```python
import statsmodels.formula.api as smf
```

```python
modelo1 = smf.ols(formula="Ingreso_deflactado~mujer+urbano+casado+horas_mes", data = data).fit()
```

```python
modelo1.summary()
```

Regresion con mas de 5 a√±os de experiencia y menos de 20 a√±os de exp.

```python
exp = data[(data['exp_laboral'] > 2) & (data['exp_laboral'] < 5)]
```

```python
modelo2 = smf.ols(formula="Ingreso_deflactado~mujer+urbano+casado+horas_mes", data = exp).fit()
```

```python
modelo2.summary()
```
