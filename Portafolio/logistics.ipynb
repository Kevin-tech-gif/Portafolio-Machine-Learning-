## Exportaciones

```python
exports = data.groupby('origin', as_index=False).sum()[['origin', 'total_value']]
exports.sort_values(["total_value"], ascending=False)
exports = exports.values.tolist()
exports                  
```

## Importaciones

```python
imports = data.groupby('destination', as_index=False).sum()[['destination', 'total_value']]
imports.sort_values(['total_value'], ascending=False)
```

```python
imports = imports.values.tolist()
```

```python
countries = []
for i in range(len(exports)):
    for j in range(len(imports)):
        if(exports[i][0] == imports[j][0]):
            countries.append({'country': exports[i][0], 'total_value': (exports[i][1] + imports[j][1])})

def totalValueFunc(e):
    return e["total_value"]

countries.sort(reverse=True, key=totalValueFunc)
countries

summation = 0

for i in range(len(countries)):
    summation += countries[i]['total_value']

summation 
```

```python
for i in range(len(countries)):
    countries[i]['percentage'] = (countries[i]['total_value'] / summation ) * 100
countries
```

```python
total_percentage = 0
list_countries = []

for i in range(len(countries)):
    if(total_percentage >= 80):
        break
    else:
        total_percentage += countries[i]['percentage']
        list_countries.append({'country': countries[i]['country']})
        
```
