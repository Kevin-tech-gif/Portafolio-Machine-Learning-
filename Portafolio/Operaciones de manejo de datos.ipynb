Agrupacion de datos

```python
gender = ["male", "female"]
income = ["poor", "midle class", "rich"]
```

```python
gender
```

```python
income
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,500):
        gender_data.append(np.random.choice(gender))
        income_data.append(np.random.choice(income))

```

```python
gender_data
```

```python
income_data
```

```python
altura = 160 + 30 * np.random.randn(n)
peso = 66 + 25 * np.random.randn(n)
edad = 30 + 12 * np.random.randn(n)
ingresos = 18000 + 3500 * np.random.rand(n)
```

```python
data = pd.DataFrame(
    {
       'gender': gender_data, 
       'economia status': income_data,
       'beight': altura,
       'weight': peso,
       'Age' : edad,
       'Income': ingresos
    }
)
```

```python
data.head()
```

```python
grouped_gender = data.groupby("gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
    
```
